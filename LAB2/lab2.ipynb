```python
import pandas as pd
import numpy as np
```

```python
#đọc dữ liệu từ file CSV vào một DataFrame dftuyensinh
dftuyensinh=pd.read_csv("processed_dulieuxettuyendaihoc.csv",index_col="STT")
```

```python
#sắp xếp này được thực hiện theo thứ tự tăng dần của giá trị trong cột "DH1".
dftuyensinh.sort_values(by="DH1")
```

```python
#sắp xếp này được thực hiện theo thứ tự tăng dần của giá trị trong từng cột GT và DH2
dftuyensinh.sort_values(by=["GT","DH2"])
```

```python
#định nghĩa hàm trả về giá trị trung bình của các số trong list scores khi được gọi.
def get_mean(scores):
    return sum(scores)/len(scores)
```

```python
#Tính tổng của cột "DH1" và lưu vào biến result
# In ra màn hình một dòng thông báo và sau đó in giá trị của biến result.
result=dftuyensinh["DH1"].sum()
print("\nThe result of applying the function on the dataframe: \n")
print(result)
```

```python
#tính phân vị 25 (Q1) của mảng g.
def Q1(g):
    return np.percentile(g,25)
```

```python
#tính phân vị 50 (Q2) của mảng g.
def Q2(g):
    return np.percentile(g,50)
```

```python
#tính phân vị 75 của mảng g.
def Q3(g):
    return np.percentile(g,75)
```

```python
# tính độ lệch chuẩn của mảng g.
def std(g):
    return np.std(g)
```

```python
# Tạo bảng pivot và tính các thống kê theo KT
pd.pivot_table(data=dftuyensinh,index="KT",values="DH1",aggfunc=["sum","mean","median","min","max",std,Q1,Q2,Q3])
```

```python
#Phần 1 câu 4
# Tạo bảng pivot và tính các thống kê theo KT và KV
pd.pivot_table(data=dftuyensinh,index=["KT", "KV"],values="DH1",aggfunc=["sum","mean","median","min","max",std,Q1,Q2,Q3])
```

```python
#phần 1 câu 5 
# Tạo bảng pivot và tính các thống kê theo KT, KV và DT
pd.pivot_table(data=dftuyensinh,index=["KT","KV","DT"],values="DH1",aggfunc=["sum","mean","median","min","max",std,Q1,Q2,Q3])
```

```python
#Mô tả độ tập trung và phân tán của dữ liệu T1
describe_t1 = dftuyensinh['T1'].describe()
print(describe_t1)
```

```python
#Vẽ biểu đồ Box-Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(dftuyensinh['T1'], vert=False)
plt.title('Box-Plot of T1')
plt.xlabel('T1 values')
plt.yticks([])
plt.show()
```

```python
#Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
mean_t1 = dftuyensinh['T1'].mean()
median_t1 = dftuyensinh['T1'].median()

if mean_t1 > median_t1:
    skewness = 'Phân phối lệch phải (positive skew)'
elif mean_t1 < median_t1:
    skewness = 'Phân phối lệch trái (negative skew)'
else:
    skewness = 'Phân phối đối xứng'

print(f"Mean: {mean_t1}, Median: {median_t1}")
print(f"Skewness: {skewness}")
```

```python
#Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
plt.figure(figsize=(10, 6))
plt.hist(dftuyensinh['T1'], bins=20, edgecolor='black')
plt.title('Histogram of T1')
plt.xlabel('T1 values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
from scipy.stats import skew, kurtosis

skewness_t1 = skew(dftuyensinh['T1'])
kurtosis_t1 = kurtosis(dftuyensinh['T1'])

print(f"Skewness of T1: {skewness_t1}")
print(f"Kurtosis of T1: {kurtosis_t1}")
```

```python
#Kiểm chứng phân phối chuẩn QQ-Plot
import scipy.stats as stats
import seaborn as sns

plt.figure(figsize=(8, 6))
stats.probplot(dftuyensinh['T1'], dist="norm", plot=plt)
plt.title('QQ Plot of T1')
plt.show()
```

```python
#câu 2
#Biểu đồ Box-Plot theo nhóm phanlopT1
import matplotlib.pyplot as plt
import seaborn as sns

# Vẽ biểu đồ Box-Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopt1', y='T1', data=dftuyensinh)
plt.title('Box Plot of T1 by phanlopt1')
plt.xlabel('phanlopt1')
plt.ylabel('T1 values')
plt.grid(True)
plt.show()
```

```python
#Kiểm chứng phân phối chuẩn QQ-Plot theo nhóm phanlopT1
import scipy.stats as stats

# Chia dữ liệu thành từng nhóm phân lớp
groups = dftuyensinh.groupby('phanlopt1')['T1']

plt.figure(figsize=(12, 8))

for name, group in groups:
    stats.probplot(group, dist="norm", plot=plt)

plt.title('QQ Plot of T1 by phanlopt1')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.legend(groups.groups.keys())
plt.grid(True)
plt.show()
```

```python
#Tính toán Covariance và Correlation
# Tính toán covariance
covariance = dftuyensinh['DH1'].cov(dftuyensinh['T1'])

# Tính toán correlation
correlation = dftuyensinh['DH1'].corr(dftuyensinh['T1'])

print(f"Covariance between DH1 and T1: {covariance}")
print(f"Correlation between DH1 and T1: {correlation}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(dftuyensinh['T1'], dftuyensinh['DH1'], alpha=0.6)
plt.title('Scatter Plot of DH1 vs T1')
plt.xlabel('T1')
plt.ylabel('DH1')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tính toán ma trận tương quan giữa DH1 và T1 cho từng nhóm khu vực
correlation_by_kv = dftuyensinh.groupby('KV')[['DH1', 'T1']].corr().iloc[0::2, -1].reset_index(level=1, drop=True)
print("Ma trận tương quan giữa DH1 và T1 trên từng nhóm khu vực:")
print(correlation_by_kv)

```

```python
plt.figure(figsize=(15, 10))
for i, (kv_value, corr) in enumerate(correlation_by_kv.items()):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(data=dftuyensinh[dftuyensinh['KV'] == kv_value], x='T1', y='DH1')
    plt.title(f'KV={kv_value} (Corr: {corr:.2f})')

plt.suptitle('Scatter plot of DH1 vs T1 by KV', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
