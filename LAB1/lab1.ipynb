```python
# vòng lặp qua các phần tử trong list ["1", "2", "6"] 
# và in ra chuỗi kết quả bao gồm "TBM" kết hợp với từng giá trị của i trong danh sách đó.
for i in ["1","2","6"]:
    print("TBM"+i)
```

```python
#thực hiện một vòng lặp để tính toán các giá trị mới cho các cột trong DataFrame df dựa trên các công thức phức tạp đã cho.
for i in ["1","2","6"]:
    df["TBM"+i]=(df["T"+i]*2 +df["L"+i]+ df["H"+i]+df["S"+i]+df["V"+i]*2+df["X"+i]+df["D"+i]+df["N"+i])/10
```

```python
#đổi tên của cột trong DataFrame df từ "TBM6" thành "TBM3"
df.rename(columns={"TBM6":"TBM3"},inplace=False)
```

```python
#câu 8
#hàm xep loai 
def xep_loai(tbm):
    if tbm < 5.0:
        return 'Y'
    elif tbm < 6.5:
        return 'TB'
    elif tbm < 8.0:
        return 'K'
    elif tbm < 9.0:
        return 'G'
    else:
        return 'XS'

for i in range(1, 4):
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)

# Kiểm tra kết quả
print(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']])
```

```python
#câu 9
import pandas as pd
# Min-Max Normalization function
def min_max_normalization(tbm, min_value=0, max_value=10, new_min=0, new_max=4):
    return ((tbm - min_value) / (max_value - min_value)) * (new_max - new_min) + new_min

# Tạo các biến US_TBM1, US_TBM2, US_TBM3
for i in range(1, 4):
    df[f'US_TBM{i}'] = df[f'TBM{i}'].apply(min_max_normalization)

# Kiểm tra kết quả
print(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']])

```

```python

#câu 10
def ket_qua_xt(row):
    if row['KT'] in ['A', 'A1']:
        diem_xt = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        diem_xt = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        diem_xt = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if diem_xt >= 5.0 else 0

# Tạo biến KQXT
df['KQXT'] = df.apply(ket_qua_xt, axis=1)

# Kiểm tra kết quả
print(df[['DH1', 'DH2', 'DH3', 'KT', 'KQXT']])
```
