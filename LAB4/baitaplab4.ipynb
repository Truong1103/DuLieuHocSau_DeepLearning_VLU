Câu 1 Viết hàm load_data() để tải dữ liệu lên ứng dụng. Sau đó, hiển thị ra màn hình 10 dòng đầu tiên.

```python
import pandas as pd
def load_data(file_path):
    data = pd.read_csv(file_path)
    return data
file_path = 'titanic_disaster.csv'
titanic = load_data(file_path)
```

```python
titanic.head(10)
```

Câu 2 Thống kê dữ liệu thiếu trên các biến số và trực quan hóa dữ liệu thiếu bằng biểu đồ (Heatmap)

```python
missing_data = titanic.isnull().sum()
print("Missing data statistics:")
print(missing_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values in Titanic Dataset')
plt.show()
```

Câu 3 Xử lý tên cột tên Name, tách ra làm 2 cột: firstName và secondName. Lưu ý: Sau khi
tách cột xong thì xóa luôn cột Name

```python
# Tách cột “Tên” thành “fristname” và “secondeName” dựa trên dấu phân cách bằng dấu phẩy
titanic[['Firstname', 'secondName']]=titanic['Name'].str.split(",",expand = True)
titanic.head()
```

Câu 4 Xứ lý rút gọn kích thước dữ liệu trên cột Sex như sau: thay thế male  M và female  F

```python
# Thay thế “male” bằng “M” và “female” bằng “F” trong cột “Giới tính”
titanic['Sex'].replace({'male':'M','female':'F'},inplace=True)
titanic.head(10)
```

Câu 5 Xử lý dữ liệu thiếu trên biến Age bằng cách thay thế bằng giá trị trung bình tuổi: Hãy đưa
ra quyết định dùng giá trị trung bình tuổi toàn bộ hành khách hay theo từng nhóm hạng vé (hạng hành khách: Pclass)

```python
import seaborn as sns
sns.boxplot(y='Age', x='Pclass', data = titanic)
```

```python
# Thay thế các giá trị bị thiếu trong cột 'Age' bằng giá trị trung bình
titanic['Age'].fillna(round(titanic['Age'].mean()), inplace= True)
```

```python
titanic.head(10)
```

```python
# Vẽ biểu đồ Heat map để trực quan hóa các giá trị thiếu trong DataFrame
plt.figure(figsize=(12, 8))
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap')
plt.show()
```

Câu 6 Xây dựng biến số Agegroup có thang đo thứ tự được ánh xạ theo thang đo khoảng dựa trên độ tuổi của hành khách như sau: (age =< 12] -> Kid; (12, 18]: Teen, (18, 60]: Adult và ( age > 60): Older

```python
titanic.loc[(titanic['Age'] <= 12), 'Agegroup'] = 'Kid'
titanic.loc[(titanic['Age'] > 12) & (titanic['Age'] <= 18), 'Agegroup'] = 'Teen'
titanic.loc[(titanic['Age'] > 18) & (titanic['Age'] <= 60), 'Agegroup'] = 'Adult'
titanic.loc[(titanic['Age'] > 60), 'Agegroup'] = 'Older'
     
```

```python
titanic.head(10)
```

Câu 7 Tiến hành thêm đặc trưng về danh xưng (namePrefix) trong xã hội bằng cách tách Mr,Mrs, Miss, Master ra khỏi "secondName"

```python
titanic['namePrefix'] = titanic['secondName'].str.extract(r'\b(Mr|Mrs|Miss|Master)\b', expand=False)
```

```python
titanic.head(10)
```

Câu 8 Khai thác thêm thông tin số lượng thành viên đi theo nhóm thân quen (familySize) đối
với mỗi hành khách trên chuyến hải trình; family size = 1+ SibSp + Parch

```python
titanic['familySize'] = 1 + titanic['SibSp'] + titanic['Parch']
```

```python
titanic.head(10)
```

Câu 9 Tạo thêm đặc trưng ‗Alone‘ để xác định hành khách đi theo nhóm hay cá nhân bằng cách
dựa trên familySize như sau: Nếu familySize = 0 thì giá trị Alone = 1 và ngược lại là 0.

```python
titanic['Alone'] = titanic['familySize'].apply(lambda x: 1 if x == 1 else 0)
```

```python
titanic.head(10)
```

Câu 10 Tiến hành tách loại cabin (typeCabin) mà hành khách ở để lọc và phân tích đặc tính
cabin. Loại cabin được kí hiệu bởi chữ cái đầu tiên. Lưu ý: Đối với dữ liệu cabin bị thiếu
thì thay thế bằng ―Unknown‖

```python
titanic['typeCabin'] = titanic['Cabin'].fillna('Unknown')
typeCabin = titanic['typeCabin']
for i in range(len(typeCabin)):
  temp = str(typeCabin[i])
  if temp != 'Unknown':
    typeCabin[i] = temp[:1]
```

Câu 12 Trực quan thông tin tương quan tỉ lệ sống sót và thiệt mạng trên từng nhóm giới tính

```python
corSurvivedSex = titanic.groupby(['Survived','Sex'])[['Sex']].count().unstack()
```

Câu 13 Trực quan thông tin hành khách sống sót trên từng nhóm phân loại hành khách (Pclass).

```python
# Trực quan thông tin hành khách sống sót trên từng nhóm phân loại hành khách
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival Counts by Passenger Class')
plt.show()
```

Câu 14 . Trực quan thông tin hành khách sống sót trên từng nhóm giới tính và thang đo tuổi tác

```python
# Trực quan thông tin hành khách sống sót trên từng nhóm giới tính và thang đo tuổi tác
plt.figure(figsize=(14, 8))
sns.countplot(x='Agegroup', hue='Sex', data=titanic[titanic['Survived'] == 1], palette='viridis')
plt.title('Thông tin hành khách sống sót theo từng nhóm tuổi và giới tính')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Số lượng')
plt.legend(title='Giới tính', labels=['Nam', 'Nữ'])
plt.show()
```

Câu 15 Trực quan xác suất hành khách sống sót dựa trên thông tin nhóm đi cùng


```python
titanic['familySize'] = titanic['SibSp'] + titanic.Parch + 1
```

```python
# Sử dụng catplot để vẽ biểu đồ
sns.catplot(x='familySize', y='Survived', data=titanic, kind='point', aspect=2, height=6, palette='viridis')
plt.title('Xác suất sống sót theo kích thước nhóm đi cùng')
plt.xlabel('Kích thước nhóm đi cùng')
plt.ylabel('Xác suất sống sót')
plt.show()
```

Câu 16. Trực quan xác suất hành khách sống sót dựa trên thông tin giá vé


```python
# Sử dụng histplot để vẽ biểu đồ phân phối giá vé
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='Fare', hue='Survived', kde=True, palette='viridis', bins=30)
plt.title('Phân phối giá vé theo sống sót và không sống sót')
plt.xlabel('Giá vé')
plt.ylabel('Số lượng')
plt.legend(labels=['Không sống sót', 'Sống sót'])
plt.show()
```

Câu 17. Trực quan số lượng người thiệt mạng và sống sót theo phân lớp (Pclass) hành khách và
cảng sẽ cập bến

```python
sns.catplot(x='Pclass', y='Survived', hue='Sex', col='Embarked', data=titanic, kind='point', palette='viridis', aspect=1.5)
plt.subplots_adjust(top=0.9)
plt.suptitle('Xác suất sống sót theo phân lớp hành khách, giới tính và cảng cập bến', fontsize=16)
plt.show()
```
